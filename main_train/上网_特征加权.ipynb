```python
import missingno as msno
from time import time
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
plt.rcParams['font.sans-serif']=['SimHei']
```

```python
factor = '网络覆盖与信号强度'
```

```python
train = pd.read_excel('output/手机上网'+ factor +'.xls', index_col=0)
```

```python
train_data = pd.DataFrame.copy(train, deep=True)
train_data
```

```python
cor = pd.DataFrame(train_data.corr().abs()[factor].sort_values(ascending = False))
cor.columns = [factor]
cor
```

```python
corr1 = cor[1:]
corr1
```

```python
sum1 = corr1.apply(sum)[0]
sum1
```

```python
corr1 = corr1.apply(lambda a:a/sum1)
corr1
```

```python
corr1[factor]['手机上网速度慢']
```

```python
corr1.to_excel('output/corr/手机上网'+factor+'.xlsx')
```

```python
titles = corr1.index.tolist()
titles
```

```python
for title in titles:
    train_data[title] = train_data[title].apply(lambda a : a*corr1[factor][title])
train_data
```
